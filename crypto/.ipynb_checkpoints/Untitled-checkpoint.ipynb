```python
def Prime_less_root(n):
    A = [True] * (n + 1)
    
    for i in range(2, n + 1):
        if A[i]:
            for j in range(i * i, n + 1, i):
                A[j] = False

    return [i for i in range(2, n + 1) if A[i]]
```

```python
def erastothene_sieve(n):
    R = n**(1/2)
    A = Prime_less_root(int(R))

    for i in range(1, len(A)):
        if n % A[i] == 0:
            return False
    return True
```

```python
def Extended_Euclidean(a,b):
    old_r = a
    old_u = 1
    old_v = 0
    r = b
    u = 0
    v = 1

    while r != 0:
        quotient = old_r // r
        old_r, r = r, old_r - quotient * r
        old_u, u = u, old_u - quotient * u
        old_v, v = v, old_v - quotient * v
    # r = old_r - quotient * r
    # s = old_s - quotient * s
    # t = old_t - quotient * t
    d = old_r
    u = old_u
    v = v


    return u, v, d
```

```python
from sage.parallel.decorate import parallel
import random

def pseudoprimality_test(n):
    """
    Performs a single pseudoprimality test on n.

    Args:
        n: The integer to test.

    Returns:
        "composite" if n is definitely composite, "probably prime" otherwise.
    """
    if n <= 1:
        return "composite"
    
    F = factorint(n - 1)
    p = list(F.keys())
    k = list(F.values())

    e = 0
    m = 1
    for i in range(len(p)):
        if p[i] == 2:
            e = k[i]
        m *= p[i] ** k[i]

    x = random.randint(1, n - 1)
    d = gcd(x, m)  # Use built-in gcd
    if d != 1:
        return "composite"
    
    y = power_mod(x, m, n)
    d = gcd(y, n)  # Use built-in gcd
    if d == 1:
        return "probably prime"
    
    for i in range(e):
        y = power_mod(x, m * (2 ** i), n)
        d = gcd(y, n) # Use built-in gcd
        if d == -1:
            return "probably prime"
        else:
            y = (y * y) % n
            
    return "composite"

@parallel()
def pseudoprimality_test_parallel(n):
    return pseudoprimality_test(n)


def strong_pseudoprimality_test(n, t):
    """
    Performs t pseudoprimality tests on n and returns "composite" if any test
    returns "composite", otherwise returns "probably prime".  Parallelized.

    Args:
        n: The integer to test.
        t: The number of tests to perform.

    Returns:
        "composite" if n is composite, "probably prime" otherwise.
    """
    
    # Parallelize the pseudoprimality tests
    results = pseudoprimality_test_parallel([n] * t) # Create a list of 'n' repeated t times
    
    if "composite" in results:
        return "composite"
    return "probably prime"


def generate_safe_prime(n):
    """
    Generates a safe prime (p such that (p-1)/2 is also prime) with approximately n bits.

    Args:
        n: The desired number of bits in the safe prime.

    Returns:
        A safe prime.
    """
    x = 2**n
    y0 = (x - (log(x))^2) / (2 * (log(x))^2)
    y1 = 2**(n - 1) - 1
    t = 10

    while True:
        l = random.randint(int(y0), int(y1))
        if (strong_pseudoprimality_test(l, t) == "probably prime" and
            strong_pseudoprimality_test(2 * l + 1, t) == "probably prime"):
            return 2 * l + 1

        for j in range(int(n/4), int(n/2)):
            l1_min = 2**j + 1
            l1_max = 2**(j + 1)

            if l1_min >= l1_max:
                continue

            l1 = random.randint(l1_min, l1_max - 1)

            l2_min = y0 / l1
            l2_max = y1 / l1

            if l2_min >= l2_max:
                continue

            l2 = random.uniform(l2_min, l2_max)
            l2 = int(l2)

            if (strong_pseudoprimality_test(l1, t) == "probably prime" and
                strong_pseudoprimality_test(l2, t) == "probably prime" and
                strong_pseudoprimality_test(2 * l1 * l2 + 1, t) == "probably prime"):
                return 2 * l1 * l2 + 1

# Example usage
safe_prime = generate_safe_prime(30)
print(safe_prime)
```

```python
from sage.parallel.decorate import parallel
import random

def pseudoprimality_test(n):
    """
    Performs a single pseudoprimality test on n.

    Args:
        n: The integer to test.

    Returns:
        "composite" if n is definitely composite, "probably prime" otherwise.
    """
    if n <= 1:
        return "composite"
    
    F = factorint(n - 1)
    p = list(F.keys())
    k = list(F.values())

    e = 0
    m = 1
    for i in range(len(p)):
        if p[i] == 2:
            e = k[i]
        m *= p[i] ** k[i]

    x = random.randint(1, n - 1)
    d = gcd(x, m)  # Use built-in gcd
    if d != 1:
        return "composite"
    
    y = power_mod(x, m, n)
    d = gcd(y, n)  # Use built-in gcd
    if d == 1:
        return "probably prime"
    
    for i in range(e):
        y = power_mod(x, m * (2 ** i), n)
        d = gcd(y, n) # Use built-in gcd
        if d == -1:
            return "probably prime"
        else:
            y = (y * y) % n
            
    return "composite"

@parallel()
def pseudoprimality_test_parallel(n):
    return pseudoprimality_test(n)


def strong_pseudoprimality_test(n, t):
    """
    Performs t pseudoprimality tests on n and returns "composite" if any test
    returns "composite", otherwise returns "probably prime".  Parallelized.

    Args:
        n: The integer to test.
        t: The number of tests to perform.

    Returns:
        "composite" if n is composite, "probably prime" otherwise.
    """
    
    # Parallelize the pseudoprimality tests
    results = pseudoprimality_test_parallel([n] * t) # Create a list of 'n' repeated t times
    
    if "composite" in results:
        return "composite"
    return "probably prime"


def generate_safe_prime(n):
    """
    Generates a safe prime (p such that (p-1)/2 is also prime) with approximately n bits.

    Args:
        n: The desired number of bits in the safe prime.

    Returns:
        A safe prime.
    """
    x = 2**n
    y0 = (x - (log(x))^2) / (2 * (log(x))^2)
    y1 = 2**(n - 1) - 1
    t = 10

    while True:
        l = random.randint(int(y0), int(y1))
        if (strong_pseudoprimality_test(l, t) == "probably prime" and
            strong_pseudoprimality_test(2 * l + 1, t) == "probably prime"):
            return 2 * l + 1

        for j in range(int(n/4), int(n/2)):
            l1_min = 2**j + 1
            l1_max = 2**(j + 1)

            if l1_min >= l1_max:
                continue

            l1 = random.randint(l1_min, l1_max - 1)

            l2_min = y0 / l1
            l2_max = y1 / l1

            if l2_min >= l2_max:
                continue

            l2 = random.uniform(l2_min, l2_max)
            l2 = int(l2)

            if (strong_pseudoprimality_test(l1, t) == "probably prime" and
                strong_pseudoprimality_test(l2, t) == "probably prime" and
                strong_pseudoprimality_test(2 * l1 * l2 + 1, t) == "probably prime"):
                return 2 * l1 * l2 + 1

def generate_rsa_keypair(num_digits):
    """
    Generates an RSA key pair (public key, private key).

    Args:
        num_digits: The approximate number of digits for each prime number.

    Returns:
        A tuple: (public_key, private_key), where public_key is (n, e) and
        private_key is (n, d).
    """
    p = generate_safe_prime(num_digits)
    q = generate_safe_prime(num_digits)

    n = p * q
    phi = (p - 1) * (q - 1)

    # Choose an e such that 1 < e < phi and gcd(e, phi) = 1
    e = 65537  # A common choice for e
    while gcd(e, phi) != 1:
        e = random.randint(2, phi - 1)  # If 65537 doesn't work find another e.

    # Find the modular inverse of e modulo phi (private key)
    d = inverse_mod(e, phi)

    public_key = (n, e)
    private_key = (n, d)

    return public_key, private_key


def encrypt(message, public_key):
    """
    Encrypts a message using the public key.

    Args:
        message: The message to encrypt (an integer).
        public_key: The recipient's public key (n, e).

    Returns:
        The encrypted message (ciphertext), an integer.
    """
    n, e = public_key
    ciphertext = power_mod(message, e, n)
    return ciphertext


def decrypt(ciphertext, private_key):
    """
    Decrypts a ciphertext using the private key.

    Args:
        ciphertext: The ciphertext to decrypt (an integer).
        private_key: The recipient's private key (n, d).

    Returns:
        The decrypted message (plaintext), an integer.
    """
    n, d = private_key
    plaintext = power_mod(ciphertext, d, n)
    return plaintext


# ---- Main Execution ----
if __name__ == '__main__':
    num_digits = 1990

    # Generate key pair
    public_key, private_key = generate_rsa_keypair(num_digits)

    print("Public key (n, e):", public_key)
    print("Private key (n, d):", private_key)

    # Example: Encrypt and decrypt a message
    message = 1234567890  # Replace with a meaningful integer message

    # Ensure the message is smaller than n:
    n = public_key[0]
    if message >= n:
        message = message % (n // 1000) # Reduce message size

    ciphertext = encrypt(message, public_key)
    print("Ciphertext:", ciphertext)

    plaintext = decrypt(ciphertext, private_key)
    print("Plaintext:", plaintext)

    if message == plaintext:
        print("Encryption and decryption successful!")
    else:
        print("Encryption and decryption failed.")
```
