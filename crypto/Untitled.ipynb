```python
def Prime_less_root(n):
    A = [True] * (n + 1)
    
    for i in range(2, n + 1):
        if A[i]:
            for j in range(i * i, n + 1, i):
                A[j] = False

    return [i for i in range(2, n + 1) if A[i]]
```

```python
def erastothene_sieve(n):
    R = n**(1/2)
    A = Prime_less_root(int(R))

    for i in range(1, len(A)):
        if n % A[i] == 0:
            return False
    return True
```

```python
def Extended_Euclidean(a,b):
    old_r = a
    old_u = 1
    old_v = 0
    r = b
    u = 0
    v = 1

    while r != 0:
        quotient = old_r // r
        old_r, r = r, old_r - quotient * r
        old_u, u = u, old_u - quotient * u
        old_v, v = v, old_v - quotient * v
    # r = old_r - quotient * r
    # s = old_s - quotient * s
    # t = old_t - quotient * t
    d = old_r
    u = old_u
    v = v


    return u, v, d
```

```python
from sage.parallel.decorate import parallel
import random
from sage.crypto.util import random_prime

def is_prime(n, k=5):  # Miller-Rabin primality test
    """
    Miller-Rabin primality test.

    Args:
        n: The number to test for primality.
        k: The number of rounds of testing to perform (higher k means higher confidence).

    Returns:
        True if n is probably prime, False if n is composite.
    """
    if n <= 1:
        return False
    if n <= 3:
        return True

    # Find r and d such that n-1 = 2^r * d
    r, d = 0, n - 1
    while d % 2 == 0:
        r += 1
        d //= 2

    for _ in range(k):
        a = random.randint(2, n - 2)
        x = power_mod(a, d, n)

        if x == 1 or x == n - 1:
            continue

        for _ in range(r - 1):
            x = power_mod(x, 2, n)
            if x == n - 1:
                break
        else:
            return False  # Composite

    return True  # Probably prime



def generate_safe_prime(n):
    """
    Generates a safe prime (p such that (p-1)/2 is also prime) with approximately n bits.

    Args:
        n: The desired number of bits in the safe prime.

    Returns:
        A safe prime.
    """
    while True:
        q = random_prime(2**(n-1), 2**n)  # Generate a random prime q with n-1 to n bits
        p = 2 * q + 1
        if is_prime(p): # or p.is_prime() in Sage
            return p

# Example usage
safe_prime = generate_safe_prime(30)
print(safe_prime)
print(f"Is the generated number prime?: {is_prime(safe_prime)}")
print(f"Is (safe_prime-1)/2 prime?: {is_prime((safe_prime-1)//2)}")
```

```python
from sage.all import randint, is_prime, gcd, log, power_mod

def strong_pseudoprimality_test(N):
    """
    Strong pseudoprimality test based on the algorithm provided.

    Args:
        N: An odd integer to test (N >= 3).

    Returns:
        "composite" if N is composite, "probably prime" otherwise.
    """
    if N < 3 or N % 2 == 0:
        return "composite"

    # Step 1: Write N - 1 = 2^e * m
    e = 0
    m = N - 1
    while m % 2 == 0:
        e += 1
        m //= 2

    # Step 2: Choose x uniformly at random from {1, ..., N - 1}
    x = randint(2, N - 2)

    # Step 3: If gcd(x, N) != 1 then return "composite"
    if gcd(x, N) != 1:
        return "composite"

    # Step 4: Compute y = x^m mod N
    y = power_mod(x, m, N)

    # Step 5: If y == 1, return "probably prime"
    if y == 1:
        return "probably prime"

    # Step 6: Loop over i from 0 to e - 1
    for _ in range(e):
        # Step 7: If y == N - 1, return "probably prime"
        if y == N - 1:
            return "probably prime"
        # Step 8: y = y^2 mod N
        y = power_mod(y, 2, N)

    # Step 9: If none of the conditions are met, return "composite"
    return "composite"

def generate_safe_prime(n):
    """
    Generates a safe prime (p such that (p-1)/2 is also prime) with approximately n bits.

    Args:
        n: The desired number of bits in the safe prime.

    Returns:
        A safe prime.
    """
    x = 2**n
    y0 = (x - log(x, 2)) / (2 * log(x, 2))
    y1 = 2**(n - 1) - 1

    while True:
        # Step 3: Choose a uniformly random integer l
        l = randint(int(y0), int(y1))
        if (strong_pseudoprimality_test(l) == "probably prime" and
            strong_pseudoprimality_test(2 * l + 1) == "probably prime"):
            return 2 * l + 1

        # Steps 4-6: Additional generation if needed
        for j in range(int(n / 4), int(n / 2)):
            l1_min = 2**j + 1
            l1_max = 2**(j + 1)

            if l1_min >= l1_max:
                continue

            l1 = randint(l1_min, l1_max - 1)
            l2_min = y0 / l1
            l2_max = y1 / l1

            if l2_min >= l2_max:
                continue

            l2 = randint(int(l2_min), int(l2_max))

            if (strong_pseudoprimality_test(l1) == "probably prime" and
                strong_pseudoprimality_test(l2) == "probably prime" and
                strong_pseudoprimality_test(2 * l1 * l2 + 1) == "probably prime"):
                return 2 * l1 * l2 + 1

# Example usage
safe_prime = generate_safe_prime(30)
print(safe_prime)
```

```python
%%time

from sage.all import randint, is_prime, gcd, log, power_mod
from sage.parallel.decorate import parallel

@parallel()
def strong_pseudoprimality_test_parallel(N):
    """
    Strong pseudoprimality test (Miller-Rabin) that can be called in parallel.
    """
    if N < 3 or N % 2 == 0:
        return "composite"
    e, m = 0, N - 1
    while m % 2 == 0:
        e += 1
        m //= 2
    x = randint(2, N - 2)
    if gcd(x, N) != 1:
        return "composite"
    y = power_mod(x, m, N)
    if y == 1:
        return "probably prime"
    for _ in range(e):
        if y == N - 1:
            return "probably prime"
        y = power_mod(y, 2, N)
    return "composite"

def generate_safe_prime(n, num_tests=10):
    """
    Generates a probable safe prime (p such that (p-1)/2 is also prime) with approximately n bits,
    using parallel strong pseudoprimality tests.
    """
    x = 2**n
    y0 = (x - log(x, 2)) / (2 * log(x, 2))
    y1 = 2**(n - 1) - 1

    while True:
        l = randint(int(y0), int(y1))

        # Parallelize the pseudoprimality tests for l and 2*l+1
        results = strong_pseudoprimality_test_parallel([l, 2*l + 1])
        if all(r == "probably prime" for r in results):
            return 2 * l + 1

        for j in range(int(n / 4), int(n / 2)):
            l1_min = 2**j + 1
            l1_max = 2**(j + 1)

            if l1_min >= l1_max:
                continue

            l1 = randint(l1_min, l1_max - 1)
            l2_min = y0 / l1
            l2_max = y1 / l1

            if l2_min >= l2_max:
                continue

            l2 = randint(int(l2_min), int(l2_max))

            # Parallelize the pseudoprimality tests for l1, l2, and 2*l1*l2+1
            results = strong_pseudoprimality_test_parallel([l1, l2, 2 * l1 * l2 + 1])
            if all(r == "probably prime" for r in results):
                return 2 * l1 * l2 + 1
```

```python
from sage.parallel.decorate import parallel
import random

def pseudoprimality_test(n):
    """
    Performs a single Miller-Rabin pseudoprimality test on n with a randomly chosen base.
    This is a probabilistic primality test and *not* guaranteed to be accurate, but it is
    significantly more reliable than the Fermat test. Use with caution.

    Args:
        n: The integer to test.

    Returns:
        "composite" if n is definitely composite, "probably prime" otherwise.
    """
    if n <= 1:
        return "composite"
    if n <= 3:
        return "probably prime"

    # Find r and d such that n-1 = 2^r * d
    r, d = 0, n - 1
    while d % 2 == 0:
        r += 1
        d //= 2

    # Choose a random base a in the range [2, n-2]
    a = random.randint(2, n - 2)

    x = power_mod(a, d, n)
    if x == 1 or x == n - 1:
        return "probably prime"

    for _ in range(r - 1):
        x = power_mod(x, 2, n)
        if x == n - 1:
            return "probably prime"

    return "composite"  # Composite


@parallel()
def pseudoprimality_test_parallel(n):
    return pseudoprimality_test(n)


def strong_pseudoprimality_test(n, t):
    """
    Performs t pseudoprimality tests on n and returns "composite" if any test
    returns "composite", otherwise returns "probably prime".  Parallelized.

    Args:
        n: The integer to test.
        t: The number of tests to perform.

    Returns:
        "composite" if n is composite, "probably prime otherwise.
    """

    # Parallelize the pseudoprimality tests
    results = pseudoprimality_test_parallel([n] * t)  # Create a list of 'n' repeated t times

    if "composite" in results:
        return "composite"
    return "probably prime"


def generate_safe_prime(n, tests=10):
    """
    Generates a safe prime (p such that (p-1)/2 is also prime) with approximately n bits,
    using the Miller-Rabin pseudoprimality test. While significantly better than using Fermat,
    this *is still not guaranteed* to produce a true safe prime, as it relies on a
    probabilistic primality test. Use with caution, and *verify* the results if safety is critical.

    Args:
        n: The desired number of bits in the safe prime.
        tests: The number of pseudoprimality tests to run on both q and 2q+1.

    Returns:
        A probable safe prime (may be composite).
    """
    while True:
        q = random.randint(2**(n-1), 2**n - 1)
        if strong_pseudoprimality_test(q, tests) == "probably prime":  # Check if q is probably prime
            p = 2 * q + 1
            if strong_pseudoprimality_test(p, tests) == "probably prime":  # Check if 2q+1 is probably prime
                return p


# Example usage
safe_prime = generate_safe_prime(30)
print(safe_prime)

# Verify if the result is actually prime (for testing/debugging purposes):
print(f"Is the generated number likely prime (Miller-Rabin)? Result: {strong_pseudoprimality_test(safe_prime, 20)}")
print(f"Is (safe_prime-1)/2 likely prime (Miller-Rabin)? Result: {strong_pseudoprimality_test((safe_prime-1)//2, 20)}")

# To be REALLY SURE, use the built-in is_prime:
# print(f"Is the generated number really prime? Result: {is_prime(safe_prime)}")
# print(f"Is (safe_prime-1)/2 really prime? Result: {is_prime((safe_prime-1)//2)}")
```

```python
from sage.all import randint, is_prime, gcd, log, power_mod
from sympy import factorint

def iis_prime(n):
    """
    Check if a number is prime using trial division.
    """
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i = i + 6
    return True

def erastothene_sieve(n):
    """
   This function attempts to check if a number n is "prime"
    

    Args:
      n: The integer to test.

    Returns:
      False.  This will almost always return false becasue the numbers 1 up to the sqrt of n
      contains the number one which isn't prime, and since the first check is if numbers from 1 up to the
      sqrt of n is prime, the function will automatically return false.

    """
    R = int(n**(1/2))
    #A = Prime_less_root(int(R))

    for i in range(1, R+1):
        if iis_prime(i) == False:
            return False
        if n % i == 0:
            return False
    return True

def Extended_Euclidean(a, b):
    """
    Implements the Extended Euclidean Algorithm to find the greatest common divisor (gcd)
    of two integers a and b, as well as the coefficients u and v such that au + bv = gcd(a, b).

    Args:
      a: The first integer.
      b: The second integer.

    Returns:
      A tuple (u, v, d) where:
        u: The coefficient for a.
        v: The coefficient for b.
        d: The greatest common divisor of a and b.
    """
    old_r = a
    old_u = 1
    old_v = 0
    r = b
    u = 0
    v = 1

    while r != 0:
        quotient = old_r // r
        old_r, r = r, old_r - quotient * r
        old_u, u = u, old_u - quotient * u
        old_v, v = v, old_v - quotient * v

    d = old_r
    u = old_u
    v = v

    return u, v, d

def Euler_totient(n):
    """
    Calculates Euler's totient function (phi) for a given integer n.
    phi(n) is the number of integers between 1 and n (inclusive) that are
    relatively prime to n (i.e., their greatest common divisor is 1).

    Args:
      n: An integer greater than 1.

    Returns:
      The value of Euler's totient function for n.
    """
    if n <= 1:
        return 0
    if erastothene_sieve(n):
        return n - 1  # If n is prime, phi(n) = n - 1

    factors = factorint(n)
    if len(factors) == 1:
        # If n is a power of a single prime (n = p^k)
        p, k = list(factors.items())[0]
        return p ** k - p ** (k - 1)
    else:
        # If n has multiple prime factors
        phi = 1
        for p, k in factors.items():
            phi *= (p ** k - p**(k-1))
        return phi

def inverse(a, n):
    """
    Calculates the modular multiplicative inverse of a modulo n.

    Args:
      a: The integer for which to find the inverse.
      n: The modulus.

    Returns:
      The modular multiplicative inverse of a modulo n, or "a is not invertible in Z/nZ" if
      the inverse does not exist (i.e., a and n are not relatively prime).
    """
    u, v, d = Extended_Euclidean(a, n)
    if d != 1:
        return "a is not invertible in Z/nZ"
    return u % n

def decrypted(e, n, M):
    """
    Decrypts a message M 

    Args:
      e: The public exponent.
      n: The modulus (product of two primes).
      M: The ciphertext message (an integer).

    Returns:
      The decrypted plaintext message (an integer). Returns nothing if decryption fails.
    """
    phi = Euler_totient(n)
    u, v, d = Extended_Euclidean(e, phi)
    if d!= 1:
        return  # return nothing if inverse doesn't exist
    secret_key = inverse(e, phi)
    decrypt = power_mod(M, secret_key, n)
    return decrypt

def numbers_to_letters(input_string):
    """Converts numbers (00-36) to letters (Space, A-Z)."""
    letter_map = {
        "00": " ",
        "11": "A", "12": "B", "13": "C", "14": "D", "15": "E", "16": "F", "17": "G",
        "18": "H", "19": "I", "20": "J", "21": "K", "22": "L", "23": "M", "24": "N",
        "25": "O", "26": "P", "27": "Q", "28": "R", "29": "S", "30": "T", "31": "U",
        "32": "V", "33": "W", "34": "X", "35": "Y", "36": "Z"
    }
    message = ""
    input_string = str(input_string)
    for i in range(0, len(input_string), 2):
        pair = input_string[i:i+2]
        #print(f"Processing pair: {pair}")

        for key, value in letter_map.items():
            if key == pair:
                message += value
                break
    return message

```
