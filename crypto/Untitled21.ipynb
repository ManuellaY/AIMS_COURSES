```python
import random

def Extended_Euclidean(a, b):
    """
    Extended Euclidean Algorithm (iterative version).
    Returns u, v, d such that au + bv = d = gcd(a, b).
    """
    old_r = a
    old_u = 1
    old_v = 0
    r = b
    u = 0
    v = 1

    while r != 0:
        quotient = old_r // r
        old_r, r = r, old_r - quotient * r
        old_u, u = u, old_u - quotient * u
        old_v, v = v, old_v - quotient * v

    d = old_r
    u = old_u
    v = old_v

    return u, v, d

def gcd(a, b):
    """Compute the greatest common divisor of a and b using extended Euclidean"""
    u, v, d = Extended_Euclidean(a, b)
    return d

def strong_pseudoprime_test(N, a):
    """
    Algorithm 3.7. Strong pseudoprimality test.
    Input: An odd integer N ∈ Z with N ≥ 3, and a base a.
    Output: Either “composite” or “probably prime”.
    """
    # 1. Write N − 1 = 2^e * m, where e and m are (uniquely determined) positive
    # integers with m odd.
    e = 0
    m = N - 1
    while m % 2 == 0:
        e += 1
        m //= 2

    # 2.  y ← a^m mod N
    y = pow(a, m, N)

    # 3. If y ≡ 1 (mod N), then return “probably prime”
    if y == 1:
        return "probably prime"

    # 4. For i from 0 to e − 1 do steps 5–6
    for _ in range(e):
        # 5. If y ≡ −1 (mod N), then return “probably prime”
        if y == N - 1:
            return "probably prime"
        # 6. y ← y^2 mod N
        y = (y * y) % N

    # 7. Return “composite”
    return "composite"

def is_prime(N, k=5):
    """
    Primality test using strong pseudoprime test.  Tests k random bases.
    """
    if N <= 1:
        return False
    if N <= 3:
        return True
    if N % 2 == 0:
        return False

    for _ in range(k):
        a = random.randint(2, N - 2)
        if gcd(a, N) != 1:
            return False # composite
        if strong_pseudoprime_test(N, a) == "composite":
            return False # composite
    return True # probably prime


def generate_large_prime(digits):
    while True:
        n = random.randint(10**(digits-1), 10**digits - 1)
        if is_prime(n):
            return n

def mod_inverse(a, m):
    u, v, d = Extended_Euclidean(a, m)
    if d != 1:
        raise ValueError("a and m are not coprime, no inverse exists")  # Handle non-coprime case
    return (u % m + m) % m  # Ensure a positive result

# Generate p1, p2, p
p1 = generate_large_prime(349)
p2 = generate_large_prime(350)
while p1 >= p2:
    p2 = generate_large_prime(350)
p = 2 * p1 * p2 + 1
while not is_prime(p) or len(str(p)) < 700:
    p1 = generate_large_prime(349)
    p2 = generate_large_prime(350)
    while p1 >= p2:
        p2 = generate_large_prime(350)
    p = 2 * p1 * p2 + 1

print(f"p1 digits: {len(str(p1))}")
print(f"p2 digits: {len(str(p2))}")
print(f"p digits: {len(str(p))}")

# ElGamal
g = 2
b = random.randint(10**698, 10**699 - 1)
B = pow(g, b, p)

encoding = {11: 'A', 12: 'B', 13: 'C', 14: 'D', 15: 'E', 16: 'F', 17: 'G', 18: 'H', 19: 'I',
            20: 'J', 21: 'K', 22: 'L', 23: 'M', 24: 'N', 25: 'O', 26: 'P', 27: 'Q', 28: 'R',
            29: 'S', 30: 'T', 31: 'U', 32: 'V', 33: 'W', 34: 'X', 35: 'Y', 36: 'Z', 41: ' '}

# Encrypt "HI"
message = "HI"
plaintext = [18, 19]
ciphertext = []
for x in plaintext:
    a = random.randint(10**698, 10**699 - 1)
    A = pow(g, a, p)
    k = pow(B, a, p)
    y = (x * k) % p
    ciphertext.append((y, A))
    print(f"Encrypt {x}: (y, A) = ({y}, {A})")

# Decrypt
decrypted = []
for y, A in ciphertext:
    k = pow(A, b, p)
    k_inv = mod_inverse(k, p)
    z = (y * k_inv) % p
    decrypted.append(z)
    print(f"Decrypt (y, A) = ({y}, {A}): z = {z}")

message_decoded = ''.join(encoding[z] for z in decrypted)
print(f"Original: {message}")
print(f"Decrypted: {message_decoded}")
```

```python
import random

def extended_gcd(a, b):
    """
    Extended Euclidean Algorithm.
    Returns gcd(a, b), x, y such that ax + by = gcd(a, b).
    """
    if a == 0:
        return b, 0, 1
    gcd, u1, v1 = extended_gcd(b % a, a)
    u = v1 - (b // a) * u1
    v = u1
    return gcd, u, v

def gcd(a, b):
    """Compute the greatest common divisor of a and b using extended Euclidean"""
    gcd_val, _, _ = extended_gcd(a, b)
    return gcd_val

def strong_pseudoprime_test(N, a):
    """
    Algorithm 3.7. Strong pseudoprimality test.
    Input: An odd integer N ∈ Z with N ≥ 3, and a base a.
    Output: Either “composite” or “probably prime”.
    """
    # 1. Write N − 1 = 2^e * m, where e and m are (uniquely determined) positive
    # integers with m odd.
    e = 0
    m = N - 1
    while m % 2 == 0:
        e += 1
        m //= 2

    # 2.  y ← a^m mod N
    y = pow(a, m, N)

    # 3. If y ≡ 1 (mod N), then return “probably prime”
    if y == 1:
        return "probably prime"

    # 4. For i from 0 to e − 1 do steps 5–6
    for _ in range(e):
        # 5. If y ≡ −1 (mod N), then return “probably prime”
        if y == N - 1:
            return "probably prime"
        # 6. y ← y^2 mod N
        y = (y * y) % N

    # 7. Return “composite”
    return "composite"

def is_prime(N, k=5):
    """
    Primality test using strong pseudoprime test.  Tests k random bases.
    """
    if N <= 1:
        return False
    if N <= 3:
        return True
    if N % 2 == 0:
        return False

    for _ in range(k):
        a = random.randint(2, N - 2)
        if gcd(a, N) != 1:
            return False # composite
        if strong_pseudoprime_test(N, a) == "composite":
            return False # composite
    return True # probably prime


def generate_large_prime(digits):
    while True:
        n = random.randint(10**(digits-1), 10**digits - 1)
        if is_prime(n):
            return n

def mod_inverse(a, m):
    def extended_gcd(a, b):
        if a == 0:
            return b, 0, 1
        gcd, x1, y1 = extended_gcd(b % a, a)
        x = y1 - (b // a) * x1
        y = x1
        return gcd, x, y
    _, x, _ = extended_gcd(a, m)
    return (x % m + m) % m

```

```python
%%time

import random
import multiprocessing
from concurrent.futures import ProcessPoolExecutor
import math

# Constants
MIN_DIGITS = 2
STRONG_PRIME_TEST_ROUNDS = 5  # Adjust for desired primality certainty
NUM_SAFE_PRIME_CANDIDATES = 10 #Candidates tried by each process.  More means a higher chance to finding a prime.

def gcd(a, b):
    """Calculates the greatest common divisor using Euclid's algorithm."""
    while b:
        a, b = b, a % b
    return a

def strong_pseudoprime_test(N):
    """Algorithm 3.7. Strong pseudoprimality test."""
    if N < 3 or N % 2 == 0:
        return "composite" if N > 2 else "probably prime"

    # 1. Write N - 1 = 2^e * m, where e and m are (uniquely determined) positive
    # integers with m odd.
    e = 0
    m = N - 1
    while m % 2 == 0:
        e += 1
        m //= 2

    # 2. x <- {1, ..., N - 1}.
    x = random.randint(1, N - 1)

    # 3. If gcd(x, N) != 1 then return "composite".
    if gcd(x, N) != 1:
        return "composite"

    # 4. y <- x^m in ZN.
    y = pow(x, m, N)

    # 5. If y = 1, then return "probably prime"
    if y == 1:
        return "probably prime"

    # 6. For i from 0 to e - 1 do steps 7-8
    for i in range(e):
        # 7. If y = -1 then return "probably prime"
        if y == N - 1:  # Representing -1 in ZN
            return "probably prime"
        # 8. else y <- y^2 in ZN.
        y = (y * y) % N

    # 9. Return "composite".
    return "composite"


def is_prime(n):
    """Checks primality using multiple rounds of the strong pseudoprime test."""
    for _ in range(STRONG_PRIME_TEST_ROUNDS):
        if strong_pseudoprime_test(n) == "composite":
            return False
    return True

def generate_safe_prime(n, processes=multiprocessing.cpu_count()):
    """Algorithm 3.24. Generating a safe prime with parallel primality testing."""
    x = 2 * n
    ln2 = math.log(2)
    y0_float = (x - math.log(x, 2)) / (2 * ln2)  # Changed log base to 2
    y1_float = (2**(n - 1)) - 1
    y0 = int(y0_float)
    y1 = int(y1_float)

    # Check for empty range
    if y0 >= y1:
        print(f"Warning: Empty range for safe prime generation. y0={y0}, y1={y1}, n={n}")
        return None  # Or raise an exception, depending on how you want to handle this

    with ProcessPoolExecutor(max_workers=processes) as executor:
        while True:
            # Generate multiple candidates to test in parallel
            candidates = []
            for _ in range(NUM_SAFE_PRIME_CANDIDATES): #Increase this to improve the chance of getting a prime number
                lmbda = random.randint(y0, y1)
                candidates.append((lmbda, 2 * lmbda + 1))

                j = random.randint(n // 4, n // 2 - 1) #Used // for integer division
                lmbda1_min = 2**j
                lmbda1_max = 2**(j + 1) - 1
                lmbda1 = random.randint(lmbda1_min, lmbda1_max)

                lmbda2_min_float = y0_float / lmbda1
                lmbda2_max_float = y1_float / lmbda1

                lmbda2_min = max(int(lmbda2_min_float) , 1)
                lmbda2_max = int(lmbda2_max_float)

                if lmbda2_min > lmbda2_max:
                    continue # Skip this candidate if the range is invalid


                lmbda2 = random.randint(lmbda2_min, lmbda2_max)

                prime_candidate = 2 * lmbda1 * lmbda2 + 1
                candidates.append((lmbda1, lmbda2, prime_candidate))

            # Extract numbers to test for primality
            numbers_to_test = []
            for candidate in candidates:
                if len(candidate) == 2:
                    numbers_to_test.extend(candidate)  # Test lmbda and 2*lmbda+1
                elif len(candidate) == 3:
                    numbers_to_test.extend(candidate)  # Test lmbda1, lmbda2, and 2*lmbda1*lmbda2+1

            #Test those numbers using multiprocessing
            results = list(executor.map(is_prime, numbers_to_test))

            # Check the results to see if we found a suitable prime
            i = 0
            for candidate in candidates:
                if len(candidate) == 2:
                    lmbda, prime_candidate = candidate
                    if results[i] and results[i+1]:
                        return prime_candidate
                    i += 2
                elif len(candidate) == 3:
                    lmbda1, lmbda2, prime_candidate = candidate
                    if results[i] and results[i+1] and results[i+2]:
                        return prime_candidate
                    i += 3

def generate_elgamal_prime(min_digits, processes=multiprocessing.cpu_count()):
    """Generates a prime p for ElGamal using safe prime generation."""
    n_bits = min_digits // 2  # Adjust as needed
    p = generate_safe_prime(n_bits, processes)
    if p is None: #no number generated
        raise ValueError("Could not generate safe prime for ElGamal.")

    return p #We now only care that p is prime

def generate_prime(digits):
    """Generates a prime number with the specified number of digits.
       This is now a non-parallel version to use within parallel execution.
    """
    while True:
        candidate = random.randint(10**(digits-1), 10**digits - 1)
        if is_prime(candidate):
            return candidate

def find_generator(p):
    """
    Finds a generator 'g' for the multiplicative group of integers modulo p.
    This function assumes that p is prime.
    """
    if not is_prime(p):
        raise ValueError("p must be prime")

    phi = p - 1  # Euler's totient function for a prime is p-1

    # Find prime factors of phi
    prime_factors = set()
    n = phi
    i = 2
    while i * i <= n:
        if n % i == 0:
            prime_factors.add(i)
            while n % i == 0:
                n //= i
        i += 1
    if n > 1:
        prime_factors.add(n)

    # Test potential generators
    for g in range(2, p):  # Generators must be between 2 and p-1
        is_generator = True
        for factor in prime_factors:
            if pow(g, phi // factor, p) == 1:
                is_generator = False
                break  # g is not a generator
        if is_generator:
            return g  # g is a generator

    raise ValueError("No generator found for p")


def generate_keypair(p):
    """Generates ElGamal public/private key pair."""
    g = find_generator(p) #use generated g
    private_key = random.randint(2, p - 2)
    public_key = pow(g, private_key, p)  # h = g^private_key mod p

    return (public_key, g, p), private_key


def encrypt(message, public_key):
    """Encrypts a message using ElGamal."""
    h, g, p = public_key
    k = random.randint(2, p - 2)  # Ephemeral key
    a = pow(g, k, p)
    b = (message * pow(h, k, p)) % p  # b = m * h^k mod p
    return a, b

def decrypt(a, b, private_key, p):
    """Decrypts an ElGamal ciphertext."""
    return (b * pow(a, p - 1 - private_key, p)) % p

# Main Execution
if __name__ == "__main__":
    processes = multiprocessing.cpu_count()
    # 1. Generate ElGamal Prime
    try:
        p = generate_elgamal_prime(MIN_DIGITS, processes)
    except ValueError as e:
        print(f"Error generating ElGamal prime: {e}")
        exit()

    print(f"Generated ElGamal prime p with {len(str(p))} digits: {p}")

    # 2. Generate Keypair
    public_key, private_key = generate_keypair(p)
    h, g, _ = public_key

    print("\nElGamal Keypair:")
    print(f"Public key (h, g, p): ({h}, {g}, {p})")
    print(f"Private key: {private_key}")

    # 3. Message
    message = random.randint(1, p - 1)  # Ensure message < p

    print(f"\nOriginal Message: {message}")

    # 4. Encryption
    a, b = encrypt(message, public_key)
    print("\nEncrypted Message:")
    print(f"a: {a}")
    print(f"b: {b}")

    # 5. Decryption
    decrypted_message = decrypt(a, b, private_key, p)
    print(f"\nDecrypted Message: {decrypted_message}")

    # Verify Decryption
    if decrypted_message == message:
        print("\nDecryption Successful!")
    else:
        print("\nDecryption Failed!")
```
