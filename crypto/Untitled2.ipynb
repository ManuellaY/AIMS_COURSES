```python

def Extended_Euclidean(a, b):
    """
    Extended Euclidean Algorithm.
    Returns gcd(a, b), x, y such that ax + by = gcd(a, b).
    """
    if a == 0:
        return b, 0, 1
    gcd, u1, v1 = extended_gcd(b % a, a)
    u = v1 - (b // a) * u1
    v = u1
    return gcd, u, v
```

```python
import random
import sympy
from concurrent.futures import ThreadPoolExecutor, as_completed


def generate_safe_prime_t(n):
    """
    Generates a probable safe prime (p such that (p-1)/2 is also prime) with approximately n bits.
    """
    # Calculate limits using integer arithmetic only
    x = 2 ** n
    y0 = (x - 1) // (2 * n)  # Avoid floating-point calculations
    y1 = (2 ** (n - 1)) - 1

    with ThreadPoolExecutor() as executor:
        while True:
            l = random.randint(y0, y1)
            futures = {
                executor.submit(strong_pseudoprimality_test, l): "l",
                executor.submit(strong_pseudoprimality_test, 2 * l + 1): "2l + 1"
            }
            for future in as_completed(futures):
                result = future.result()
                if futures[future] == "l":
                    result_l = result
                else:
                    result_2l_plus_1 = result

            if result_l == "probably prime" and result_2l_plus_1 == "probably prime":
                return 2 * l + 1

            for j in range(n // 4, n // 2):
                l1_min = 2 ** j + 1
                l1_max = 2 ** (j + 1)

                if l1_min >= l1_max:
                    continue

                l1 = random.randint(l1_min, l1_max - 1)
                l2_min = (y0 + 1) // l1
                l2_max = y1 // l1

                if l2_min >= l2_max:
                    continue

                l2 = random.randint(l2_min, l2_max)

                # Use parallelism to check the primality of l1, l2, and 2*l1*l2 + 1
                futures = {
                    executor.submit(strong_pseudoprimality_test, l1): "l1",
                    executor.submit(strong_pseudoprimality_test, l2): "l2",
                    executor.submit(strong_pseudoprimality_test, 2 * l1 * l2 + 1): "2*l1*l2 + 1"
                }
                results = {}
                for future in as_completed(futures):
                    results[futures[future]] = future.result()

                if (results["l1"] == "probably prime" and
                    results["l2"] == "probably prime" and
                    results["2*l1*l2 + 1"] == "probably prime"):
                    return 2 * l1 * l2 + 1

```

```python
%%time

import random
import multiprocessing
from concurrent.futures import ProcessPoolExecutor

def sieve_of_eratosthenes(limit):
    sieve = [True] * (limit + 1)
    sieve[0] = sieve[1] = False
    for i in range(2, int(limit ** 0.5) + 1):
        if sieve[i]:
            for j in range(i * i, limit + 1, i):
                sieve[j] = False
    return [i for i in range(limit + 1) if sieve[i]]

def miller_rabin(n, k=5):
    if n == 2 or n == 3:
        return True
    if n < 2 or n % 2 == 0:
        return False
    r, s = 0, n - 1
    while s % 2 == 0:
        r += 1
        s //= 2
    for _ in range(k):
        a = random.randrange(2, n - 1)
        x = pow(a, s, n)
        if x == 1 or x == n - 1:
            continue
        for _ in range(r - 1):
            x = (x * x) % n
            if x == n - 1:
                break
        else:
            return False
    return True

def is_prime(n, k=5):
    """Wrapper for miller_rabin to conform to the process pool."""
    return miller_rabin(n, k)

def generate_candidate(digits):
    return random.randint(10**(digits-1), 10**digits - 1)

def generate_large_prime(digits, processes=multiprocessing.cpu_count()):
    """Generates a large prime number using parallel processing."""
    with ProcessPoolExecutor(max_workers=processes) as executor:
        while True:
            # Generate a batch of candidate numbers
            candidates = [generate_candidate(digits) for _ in range(processes * 10)]  # Adjust multiplier as needed
            # Use the process pool to test the candidates in parallel
            results = executor.map(is_prime, candidates)  # Use is_prime wrapper

            # Check the results
            for n, is_probable_prime in zip(candidates, results):
                if is_probable_prime:
                    return n

def mod_inverse(a, m):
    def extended_gcd(a, b):
        if a == 0:
            return b, 0, 1
        gcd, x1, y1 = extended_gcd(b % a, a)
        x = y1 - (b // a) * x1
        y = x1
        return gcd, x, y
    _, x, _ = extended_gcd(a, m)
    return (x % m + m) % m

```

```python
%%time

import random
import multiprocessing
from concurrent.futures import ProcessPoolExecutor
import math



def sieve_of_eratosthenes(limit):
    sieve = [True] * (limit + 1)
    sieve[0] = sieve[1] = False
    for i in range(2, int(limit ** 0.5) + 1):
        if sieve[i]:
            for j in range(i * i, limit + 1, i):
                sieve[j] = False
    return [i for i in range(limit + 1) if sieve[i]]

def gcd(a, b):
    """Calculates the greatest common divisor using Euclid's algorithm."""
    while b:
        a, b = b, a % b
    return a

def strong_pseudoprime_test(N):
    """Algorithm 3.7. Strong pseudoprimality test."""
    if N < 3 or N % 2 == 0:
        return "composite" if N > 2 else "probably prime"

    # 1. Write N - 1 = 2^e * m, where e and m are (uniquely determined) positive
    # integers with m odd.
    e = 0
    m = N - 1
    while m % 2 == 0:
        e += 1
        m //= 2

    # 2. x <- {1, ..., N - 1}.
    x = random.randint(1, N - 1)

    # 3. If gcd(x, N) != 1 then return "composite".
    if gcd(x, N) != 1:
        return "composite"

    # 4. y <- x^m in ZN.
    y = pow(x, m, N)

    # 5. If y = 1, then return "probably prime"
    if y == 1:
        return "probably prime"

    # 6. For i from 0 to e - 1 do steps 7-8
    for i in range(e):
        # 7. If y = -1 then return "probably prime"
        if y == N - 1:  # Representing -1 in ZN
            return "probably prime"
        # 8. else y <- y^2 in ZN.
        y = (y * y) % N

    # 9. Return "composite".
    return "composite"


def is_prime(n):
    t = 5
    for i in range(t):
        if strong_pseudoprime_test(n) == "composite":
            return False
    return True

def generate_safe_prime(n, processes=multiprocessing.cpu_count()):
    """Algorithm 3.24. Generating a safe prime with parallel primality testing."""
    x = 2 * n
    ln2 = math.log(2)
    y0 = (x - math.log(x, 2)) / (2 * ln2)  # Changed log base to 2
    y1 = (2**(n - 1)) - 1
    y0 = int(y0)
    y1 = int(y1)

    with ProcessPoolExecutor(max_workers=processes) as executor:
        while True:
            # Step 3
            lmbda = random.randint(y0, y1)
            # Use process pool for primality tests
            results = list(executor.map(is_prime, [lmbda, 2 * lmbda + 1]))
            if all(results):
                return 2 * lmbda + 1

            # Step 4
            for j in range(n // 4, n // 2): #Used // for integer division
                lmbda1_min = 2**j
                lmbda1_max = 2**(j + 1) - 1
                lmbda1 = random.randint(lmbda1_min, lmbda1_max)

                lmbda2_min = max(y0 // lmbda1, 1)
                lmbda2_max = y1 // lmbda1
                if lmbda2_min > lmbda2_max:
                    continue # Skip this iteration if the range is invalid

                lmbda2 = random.randint(lmbda2_min, lmbda2_max)
                prime_candidate = 2 * lmbda1 * lmbda2 + 1
                # Use process pool for primality tests
                results = list(executor.map(is_prime, [lmbda1, lmbda2, prime_candidate]))
                if all(results):
                    return prime_candidate


def generate_elgamal_prime(min_digits, processes=multiprocessing.cpu_count()):
    """Generates a prime p for ElGamal using safe prime generation."""
    n_bits = min_digits // 2  # Adjust as needed
    while True:
        p = generate_safe_prime(n_bits, processes)

        if len(str(p)) >= min_digits:
                p1_digits = random.randint(min_digits // 3, min_digits // 2)  # Adjust ranges as needed
                with ProcessPoolExecutor(max_workers=processes) as executor:
                    p1 = next(executor.map(generate_prime, [p1_digits]))
                p2_digits = random.randint(p1_digits + 1, min_digits // 2 + (min_digits // 10)) #Ensure p2 has more digits than p1.

                with ProcessPoolExecutor(max_workers=processes) as executor:
                    p2 = next(executor.map(generate_prime, [p2_digits]))

                if p1 >= p2: #verify that they are not the same number.
                    continue #reject and try a new combination
                if ((p - 1) //2) == p1 * p2:
                    return p, p1, p2
        #Check if Safe prime matches Elgamal criteria

def generate_prime(digits):
    """Generates a prime number with the specified number of digits.
       This is now a non-parallel version to use within parallel execution.
    """
    while True:
        candidate = random.randint(10**(digits-1), 10**digits - 1)
        if is_prime(candidate):
            return candidate


def generate_keypair(p):
    """Generates ElGamal public/private key pair."""
    g = random.randint(2, p - 2)  # Generator g
    private_key = random.randint(2, p - 2)
    public_key = pow(g, private_key, p)  # h = g^private_key mod p

    return (public_key, g, p), private_key


def encrypt(message, public_key):
    """Encrypts a message using ElGamal."""
    h, g, p = public_key
    k = random.randint(2, p - 2)  # Ephemeral key
    a = pow(g, k, p)
    b = (message * pow(h, k, p)) % p  # b = m * h^k mod p
    return a, b

def decrypt(a, b, private_key, p):
    """Decrypts an ElGamal ciphertext."""
    return (b * pow(a, p - 1 - private_key, p)) % p

```

```python
%%time

# Constants
MIN_DIGITS = 700

processes = multiprocessing.cpu_count()
# 1. Generate ElGamal Prime
p, p1, p2 = generate_elgamal_prime(MIN_DIGITS, processes)

print(f"Generated ElGamal prime p with {len(str(p))} digits.")
print(f"p = 2 * {p1} * {p2} + 1 = {p}")

# 2. Generate Keypair
public_key, private_key = generate_keypair(p)
h, g, _ = public_key

print("\nElGamal Keypair:")
print(f"Public key (h, g, p): ({h}, {g}, {p})")
print(f"Private key: {private_key}")

# 3. Message
message = random.randint(1, p - 1)  # Ensure message < p

print(f"\nOriginal Message: {message}")

# 4. Encryption
a, b = encrypt(message, public_key)
print("\nEncrypted Message:")
print(f"a: {a}")
print(f"b: {b}")

# 5. Decryption
decrypted_message = decrypt(a, b, private_key, p)
print(f"\nDecrypted Message: {decrypted_message}")

# Verify Decryption
if decrypted_message == message:
    print("\nDecryption Successful!")
else:
    print("\nDecryption Failed!")
```
